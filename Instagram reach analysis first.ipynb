# Finding the percentages of reach for each feature

```python
home_reach = (df['From Home'].sum()/df['Impressions'].sum())*100
home_reach
```

```python
hashtag_reach = (df['From Hashtags'].sum()/df['Impressions'].sum())*100
hashtag_reach
```

```python
explore_reach = (df['From Explore'].sum()/df['Impressions'].sum())*100
explore_reach
```

```python
save_reach = (df['Saves'].sum()/df['Impressions'].sum())*100
save_reach
```

```python
Comments_reach = (df['Comments'].sum()/df['Impressions'].sum())*100
Comments_reach
```

```python
share_reach = (df['Shares'].sum()/df['Impressions'].sum())*100
share_reach
```

```python
like_reach = (df['Likes'].sum()/df['Impressions'].sum())*100
like_reach
```

```python
visitors_reach = (df['Profile Visits'].sum()/df['Impressions'].sum())*100
visitors_reach
```

```python
followers_reach = (df['Follows'].sum()/df['Impressions'].sum())*100
followers_reach
```

```python
#Comparision for like share comment save follow
plt.pie(labels= ["Home","Hashtag","Explore","Save","Comment","Share",
                 "Likes","Profile Visits","Follow"],
        x= [home_reach,hashtag_reach,explore_reach,save_reach,
            Comments_reach,share_reach,like_reach, visitors_reach,
            followers_reach],autopct='%1.0f%%',
        explode = [0,0,0,0.5,0.9,0.5,0.5,0.7,0.5])
#plt.xticks(rotation=90)
plt.show()
```

Only home has major role in getting more reach
